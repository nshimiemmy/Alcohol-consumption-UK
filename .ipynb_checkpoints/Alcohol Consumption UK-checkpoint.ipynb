## 1. K-Nearest Neighbors (KNN

Training a K-Nearest Neighbors (KNN) classifier and evaluating its performance using the hold-out method.

K-Nearest Neighbors (KNN) is applied to classify the cause of death based on death counts across different age groups.

```python
#Import Necessary Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler


# Encode the 'Individual cause of death description' as the target variable
label_encoder = LabelEncoder()
dfEngland['Cause_of_Death_Encoded'] = label_encoder.fit_transform(dfEngland['Individual cause of death description'])

# Select features and target variable
features = dfEngland.iloc[:, 4:-2]  # Selecting the age group columns as features
target = dfEngland['Cause_of_Death_Encoded']

# Normalize the feature data
scaler = StandardScaler()
features_normalized = scaler.fit_transform(features) 

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features_normalized, target, test_size=0.3, random_state=5)

# Train the k-NN algorithm on the training set
k = 7
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train, y_train)

# Evaluate using hold-out method (simple train/test split)
# Get accuracy on training set
ypred_train = knn.predict(X_train)
train_accuracy = round(metrics.accuracy_score(y_train, ypred_train), 2)
print("Accuracy on training set:", train_accuracy)

# Get accuracy on test set
ypred_test = knn.predict(X_test)
test_accuracy = round(metrics.accuracy_score(y_test, ypred_test), 2)
print("Accuracy on test set:", test_accuracy)
```

The model's accuracy on the training set is 0.46, which means that 46% of the causes of death were correctly classified when the model was applied to the data it was trained on. The accuracy on the test set is 0.43, indicating that 43% of the causes of death were correctly classified by the model on new and unseen data.

#### Cross-validation on a K-Nearest Neighbors (KNN) classifier using sklearn

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

k = 5
knn = KNeighborsClassifier(n_neighbors=k)
cv_scores = cross_val_score(knn, features_normalized, target, cv=5)

avg = sum(cv_scores) / len(cv_scores)
print('Average performance:', round(avg, 2))
```

On average, the model correctly predicted the cause of death 40% of the time across all five folds.

###### Evaluate the performance of a K-Nearest Neighbors (KNN) classifier for different values of k using k-fold cross-validation

```python
import math
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt

k_vals = [1,2,5,7,15,20,30,50]
avg_vals = []
sd_vals = []

for k in k_vals:
    knn = KNeighborsClassifier(n_neighbors=k)
    cv_scores = cross_val_score(knn, features_normalized, target, cv=5)
    avg = sum(cv_scores)/len(cv_scores)
    sd = math.sqrt(sum((cv_scores-avg)**2)/(len(cv_scores)-1))
    avg_vals.append(avg)
    sd_vals.append(sd)

fig = plt.figure()
plt.errorbar(k_vals, avg_vals, yerr=sd_vals)
plt.title("k-fold cross-validation as a function of k")
plt.xlabel("k")
plt.ylabel("Mean accuracy")
plt.show()
```

The optimal k is 5 and it gives the highest average cross-validation score which is 0.4

## 2. Gaussian NaÃ¯ve Bayes classifier 

The aim of Gaussian NaÃ¯ve Bayes (GNB) classifier here is to predict a category (low, medium, high) based on a numerical feature (year).

```python
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Assuming dfEngland is a DataFrame containing your data
# Aggregate data to sum up deaths for each year
yearly_total_deaths = dfEngland.groupby('Year [note 3]')['All ages'].sum()

# Use KBinsDiscretizer to categorize total deaths into 3 bins: Low, Medium, High
bin_discretizer = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform', subsample=None)
death_categories = bin_discretizer.fit_transform(yearly_total_deaths.values.reshape(-1, 1)).flatten()

# Feature: Year
X_nb = yearly_total_deaths.index.values.reshape(-1, 1)
# Target: Death Categories
y_nb = death_categories

# Split the data into training and testing sets
X_nb_train, X_nb_test, y_nb_train, y_nb_test = train_test_split(X_nb, y_nb, test_size=0.2, random_state=42)

# Initialize and fit the Gaussian Naive Bayes classifier
gnb_classifier = GaussianNB()
gnb_classifier.fit(X_nb_train, y_nb_train)

# Predict the death categories for the test set
y_nb_pred = gnb_classifier.predict(X_nb_test)

# Calculate and print the accuracy of the model
accuracy = accuracy_score(y_nb_test, y_nb_pred)
print('Accuracy:', round(accuracy, 2))
```

The Gaussian NaÃ¯ve Bayes classifier gives 100% of accuracy rate when categorizing years based on the total number of alcohol-specific deaths in England. This suggests that the model has been able to correctly classify every year in the test set into the appropriate category of total deaths ('Low', 'Medium', or 'High') and this model overfit the classification

```python
import matplotlib.pyplot as plt
import numpy as np

# Plot the categorized data
plt.figure(figsize=(10, 5))
plt.scatter(X_nb.flatten(), yearly_total_deaths, c=y_nb, cmap='viridis', alpha=0.7)
plt.colorbar(label='Death Category (Low=0, Medium=1, High=2)')
plt.xlabel('Year')
plt.ylabel('Total Deaths')
plt.title('Yearly Alcohol-Specific Deaths Categorized')
plt.grid(True)
plt.show()
```

## 3. Decision Tree

The objective of utilizing the Decision Tree model is to create a predictive model that determines whether the number of deaths for a specific entry is higher than the median count. This is a binary classification task that distinguishes high mortality from low mortality. The model is evaluated based on its accuracy and a classification report that includes precision, recall, and F1-score breakdowns.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn import metrics

# Encode categorical variables
sex = LabelEncoder()
icd = LabelEncoder()
cause = LabelEncoder()

dfEngland['Sex_encoded'] = sex.fit_transform(dfEngland['Sex'])
dfEngland['ICD_encoded'] = icd.fit_transform(dfEngland['ICD-10 code'])
dfEngland['Cause_encoded'] = cause.fit_transform(dfEngland['Individual cause of death description'])

# Define features and target
X = dfEngland[['Year [note 3]', 'Sex_encoded', 'ICD_encoded', 'Cause_encoded']]
# Define binary target variable: whether the number of deaths exceeds the median
median_deaths = dfEngland['All ages'].median()
y = dfEngland['All ages'] > median_deaths

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Decision Tree Classifier
decision_tree = DecisionTreeClassifier(max_depth=2)

# Train the classifier
decision_tree.fit(X_train, y_train)

# Make predictions on the training and test sets
y_train_pred = decision_tree.predict(X_train)
y_test_pred = decision_tree.predict(X_test)

# Print the model accuracy
print("Accuracy on training set:", round(metrics.accuracy_score(y_train, y_train_pred), 2))
print("Accuracy on test set:", round(metrics.accuracy_score(y_test, y_test_pred), 2))

# Evaluate the model
evaluation_report = classification_report(y_test, y_test_pred)
print(evaluation_report)
```

The model has performed strongly across both the training and test datasets. The key metrics reported in the classification report are accuracy, precision, recall, and F1-score. The accuracy on the test set is 88% and the Acuuracy on training is 86% which is quite good. The precision for False is 95%, while for True it is 82%. The recall for True is 95%, which is excellent, while for False, it is 83%. The F1-score for both classes is around 0.89, indicating balanced performance.

Decision Tree Classifier seems to be performing well, with strong precision and recall, especially in identifying true positives (deaths exceeding the median). The results suggest that your model is particularly reliable and efficient at detecting higher risks, with slightly less accuracy in confirming lower risks

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier

# Initialize the Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(max_depth=2)

# Train the classifier
dt_classifier.fit(X_train, y_train)

# Make predictions
y_pred = dt_classifier.predict(X_test)


# Plot the Decision Tree
plt.figure(figsize=(20,10))
plot_tree(dt_classifier, 
          filled=True, 
          rounded=True, 
          class_names=["Below Median", "Above Median"],
          feature_names=['Year', 'Sex_encoded', 'ICD_encoded', 'Cause_encoded'],
          max_depth=3)  # Limit depth for readability; remove for full tree
plt.show()
```

## 4. Logistic Regression

The goal is to create a logistic regression model that can predict high mortality of a record based on the sex and year of alcohol-specific deaths in England.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import ConfusionMatrixDisplay



# Calculate the median threshold for defining high mortality
median_threshold = dfEngland['All ages'].median()

# Convert the 'All ages' mortality counts into a binary outcome based on the median threshold
dfEngland['High_Mortality'] = (dfEngland['All ages'] > median_threshold).astype(int)

# One-hot encode the 'Sex' variable 
encoder = OneHotEncoder(sparse_output=False) 
sex_encoded = encoder.fit_transform(dfEngland[['Sex']])

# Create a DataFrame with the encoded sex variables
encoded_sex_df = pd.DataFrame(sex_encoded, columns=['Sex_' + category for category in encoder.categories_[0]])

# Combine the encoded sex variables with the Year
X = pd.concat([dfEngland['Year [note 3]'], encoded_sex_df], axis=1)
y = dfEngland['High_Mortality']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1000)


# Define the Logistic Regression model
log_classifier = LogisticRegression()

# Train the model on the training dataset
log_classifier.fit(X_train, y_train)
# Get the accuracy over the test data 
score = log_classifier.score(X_test, y_test)
print("Accuracy:", score)

# Get the predictions from the model
y_pred = log_classifier.predict(X_test)

# Generate and plot the confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

The following is the performance of the model based on the confusion matrix:

â€¢ True Negatives (TN): 56 cases where the model correctly predicted the actual negative class (not high mortality).

â€¢ False Positives (FP): 54 cases where the model incorrectly predicted high mortality when it was not high mortality.

â€¢ False Negatives (FN): 12 cases where the model failed to predict high mortality when it was high mortality.

â€¢ True Positives (TP): 67 cases where the model correctly predicted high mortality.


This indicate that the model is better at predicting the positive class, with 67 correct predictions, but has a significant number of false positives at 54 incorrect positive predictions.


Accuracy:
According to the accuracy score, the model was able to accurately predict whether a data point belonged to the high mortality group or not based on the median threshold of 'All ages' death counts, in approximately 65.08% of the cases in the test set. Out of 189 total predictions (56+54+12+67), the model made 123 correct predictions (56 TN + 67 TP). This aligns with the accuracy metric calculated from the confusion matrix.

## 5. Linear Regression Analysis

Linear regression is applied on aggregated data and visualize the trend of alcohol-specific deaths in England over the years. 

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Assuming dfEngland is a DataFrame containing your data
# Grouping data by year and summing up total deaths
total_deaths_per_year = dfEngland.groupby('Year [note 3]')['All ages'].sum().reset_index()
total_deaths_per_year.columns = ['Year', 'Total Deaths']

# Reshape the Year data for sklearn
X = total_deaths_per_year['Year'].values.reshape(-1, 1)
y = total_deaths_per_year['Total Deaths'].values

# Initialize the Linear Regression model
model = LinearRegression()

# Fit the model
model.fit(X, y)

# Extracting model coefficients
slope = model.coef_[0]
intercept = model.intercept_

# Making predictions
predictions = model.predict(X)

# Calculate R-squared
r_squared = model.score(X, y)
print(f'R-squared: {r_squared}')

# Plotting the actual data points
plt.scatter(X, y, color='darkblue', label='Actual Data')

# Plotting the regression line
plt.plot(X, predictions, color='red', linewidth=2, label='Regression Line')

# Labels and title
plt.xlabel('Year')
plt.ylabel('Total Deaths')
plt.title('Linear Regression Analysis of Total Deaths (2001-2021)')
plt.legend()

# Display the plot
plt.show()
```

The linear regression model shows an R-squared value of approximately 0.785. This means that the model can explain about 78.5% of the variation in the total number of alcohol-specific deaths in England based on the year.
